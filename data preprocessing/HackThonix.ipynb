```python
import pandas as pd

df = pd.read_csv("twitter_dataset.csv")
print(df.head())
print(df.info())
print(df['sentiment'].value_counts())
```

```python
ndf = pd.read_csv("Twitter_Data.csv")
print(ndf.head())
print(ndf.info())
print(ndf['category'].value_counts())
```

```python
def transform_sentiment(value):
    if value == 2.00:
        return 1
    elif value == 1.00:
        return 0
    elif value == 0:
        return -1
    else:
        return 0

# Apply transformation and replace column
df["new_sentiment"] = df["sentiment"].apply(transform_sentiment)

# Drop the original sentiment column
df.drop(columns=["sentiment"], inplace=True)

print(df)
```

```python
df.drop(columns=["Unnamed: 0"], inplace=True)
print(df.head())
```

```python
ndf.head()
```

```python
ndf = ndf.rename(columns={'clean_text': 'tweet', 'category': 'new_sentiment'})
```

```python
df["new_sentiment"] = df["new_sentiment"].apply(lambda x: -1 if x == -1 else (1 if x == 1 else 0))

# Concatenate the datasets (adding df2 to df1)
merged_df = pd.concat([ndf, df], ignore_index=True)


# Display dataset shape and first few rows
print(f"Shape of merged dataset: {merged_df.shape}")
print(merged_df.head())
```

```python
merged_df.to_csv("new_merged_dataset.csv", index=False)
```
